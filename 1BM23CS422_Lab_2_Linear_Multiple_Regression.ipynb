```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt

df = pd.read_csv('canada_per_capita_income.csv')
df

# Commented out IPython magic to ensure Python compatibility.
# %matplotlib inline
plt.xlabel('year')
plt.ylabel('percapitaincome')
plt.scatter(df.year,df.percapitaincome,color='red',marker='+')

new_df = df.drop('percapitaincome',axis='columns')
new_df

percapitaincome = df.percapitaincome
percapitaincome

# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(new_df,percapitaincome)

reg.predict([[2020]])

```

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt

df = pd.read_csv('salary.csv')
df=df.ffill(axis=0)

# Commented out IPython magic to ensure Python compatibility.
# %matplotlib inline
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.scatter(df.YearsExperience,df.Salary,color='red',marker='+')

new_df = df.drop('Salary',axis='columns')
# print(new_df)
salary = df.Salary
# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(new_df,salary)

reg.predict([[12]])
```

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
df3=pd.read_csv('hiring.csv')
df3.isnull().sum()
df3.head(100)
df3.experience = df3.experience.fillna(df3["experience"].mode())
df3.head(10)
df3.experience = df3.experience.astype(str)
def convert_to_int(word):
    word_dict = {'one':1, 'two':2, 'three':3, 'four':4, 'five':5, 'six':6, 'seven':7, 'eight':8,
                'nine':9, 'ten':10, 'eleven':11, 'twelve':12, 'zero':0, 0:0}
    return word_dict[word]
df3.experience=df3.experience.apply(lambda x : convert_to_int(x))
print(df3.head(10))
df3.fillna(df3.mean(),inplace=True)
print(df3.isnull().sum())
print(df3.head(10))
reg=linear_model.LinearRegression()
reg.fit(df3.drop('salary($)',axis='columns'),df3['salary($)'])
print(reg.predict([[2,9,6]]))
print(reg.predict([[12,10,10]]))
```
